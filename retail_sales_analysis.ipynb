### **Feature Engineering**

```python
# Converting invoice date type to datetime
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
# Checking for the column type
df.info()
```

```python
# Converting customer id to int 64
df['CustomerID']=df['CustomerID'].astype('Int64')
```

```python
# Checking for the column type
df.info()
```

```python
# Revenue Feature 
df['Revenue']=df['Quantity']*df['UnitPrice']
df
```

```python
# Adding Time based Features
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['MonthName'] = df['InvoiceDate'].dt.month_name()
df['Day'] = df['InvoiceDate'].dt.day
df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()
df['Hour'] = df['InvoiceDate'].dt.hour
df['Quarter'] = df['InvoiceDate'].dt.quarter
```

```python
df.head()
```

```python
# Adding Weekend Feature 
df['Is_Weekend'] = df['InvoiceDate'].dt.weekday >= 5
```

```python
df.head()
```

```python
# Order total revenue
order_total_revenue = df.groupby('InvoiceNo')['Revenue'].sum().reset_index()
order_total_revenue.columns=['InvoiceNo','Revenue']
df = df.merge(order_total_revenue, on='InvoiceNo',how='left')
```
